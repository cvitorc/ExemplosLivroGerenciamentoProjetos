```python
import matplotlib.pyplot as plt

falhas = {'Atrasos ': 40,
          'Falhas ': 30,
          'Falta': 20,
          'Problemas no processo': 10}

falhas_ordenadas = sorted(falhas.items(), key=lambda x: x[1], reverse=True)

x_axis = [falha[0] for falha in falhas_ordenadas]
y_axis = [falha[1] for falha in falhas_ordenadas]

cumulative_percentage = 0
cumulative_percentages = []
for i in range(len(y_axis)):
    cumulative_percentage += y_axis[i]/sum(y_axis)*100
    cumulative_percentages.append(cumulative_percentage)

fig, ax1 = plt.subplots()

color = 'tab:black'
ax1.bar(x_axis, y_axis, color=color)
ax1.set_ylabel('Frequência', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:red'
ax2.plot(x_axis, cumulative_percentages, color=color, marker='o')
ax2.set_ylabel('Porcentagem acumulada', color=color)
ax2.tick_params(axis='y', labelcolor=color)

ax2.set_ylim([0, 100])
ax1.set_ylim([0, max(y_axis)*1.1])
plt.xticks(rotation=45)

plt.title('Gráfico de Pareto - Falhas nas Embarcações')
plt.show()
```
