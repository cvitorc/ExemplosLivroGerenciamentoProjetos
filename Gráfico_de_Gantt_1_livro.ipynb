```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Dados de exemplo (atividades e datas de início/término)
atividades = ['Tarefa 3', 'Tarefa 2', 'Tarefa 1']
inicio = ['2024-02-10', '2024-02-05', '2024-02-01']
termino = ['2024-02-20', '2024-02-15', '2024-02-10']

# Converter as datas para o formato datetime
inicio_dt = [mdates.datestr2num(data) for data in inicio]
termino_dt = [mdates.datestr2num(data) for data in termino]

# Criar o gráfico de Gantt
plt.figure(figsize=(10, 6))
for i, atividade in enumerate(atividades):
    plt.barh(atividade, termino_dt[i] - inicio_dt[i], left=inicio_dt[i], color='skyblue')

# Configurar o eixo x
plt.gca().xaxis_date()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xlabel('Data')

# Configurar o título e a legenda
plt.title('Gráfico de Gantt')
plt.legend(['Atividades'])

# Exibir o gráfico
plt.tight_layout()
plt.grid(True)
plt.show()
```
