```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Dados de exemplo (atividades e datas de início/término)
atividades = ['Tarefa 5', 'Tarefa 4', 'Tarefa 3', 'Tarefa 2', 'Tarefa 1']
inicio = ['2024-02-20', '2024-02-15', '2024-02-10', '2024-02-05', '2024-02-01']
termino = ['2024-02-28', '2024-02-25', '2024-02-20', '2024-02-15', '2024-02-10']

# Converter as datas para o formato datetime
inicio_dt = [mdates.datestr2num(data) for data in inicio]
termino_dt = [mdates.datestr2num(data) for data in termino]

# Cores personalizadas para cada tarefa
cores = ['skyblue', 'salmon', 'lightgreen', 'gold', 'plum']

# Criar o gráfico de Gantt
plt.figure(figsize=(12, 5))
for i, atividade in enumerate(atividades):
    plt.barh(atividade, termino_dt[i] - inicio_dt[i], left=inicio_dt[i], color=cores[i])

# Configurar o eixo x
plt.gca().xaxis_date()
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xlabel('Data')

# Configurar o eixo y
plt.ylabel('Atividades')

# Configurar o título
plt.title('Gráfico de Gantt - Planejamento de Projeto')

# Adicionar legenda
plt.legend(atividades, loc='lower right')

# Exibir o gráfico
plt.tight_layout()
plt.grid(axis='x')
plt.show()
```
