```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo (custos reais e oramento planejado)
meses = ['Janeiro', 'Fevereiro', 'Maro', 'Abril', 'Maio']
custos_reais = [50000, 52000, 51000, 53000, 54000]  # Custos reais de produo (em reais)
orcamento_planejado = [50000, 51000, 52000, 53000, 54000]  # Oramento planejado (em reais)

# Calcular a diferena entre os custos reais e o oramento planejado
diferenca_custos = [custos_reais[i] - orcamento_planejado[i] for i in range(len(meses))]

# Calcular a mdia e o desvio padro das diferenas
media = np.mean(diferenca_custos)
desvio_padrao = np.std(diferenca_custos)

# Definir os limites de controle (3 desvios padro)
limite_superior = media + 3 * desvio_padrao
limite_inferior = media - 3 * desvio_padrao

# Plotar o grfico de controle
plt.figure(figsize=(10, 6))
plt.plot(meses, diferenca_custos, marker='o', color='blue', linestyle='-')
plt.axhline(media, color='red', linestyle='--', label='Mdia')
plt.axhline(limite_superior, color='green', linestyle='--', label='Limite Superior')
plt.axhline(limite_inferior, color='green', linestyle='--', label='Limite Inferior')
plt.xlabel('Ms')
plt.ylabel('Diferena de Custos (R$)')
plt.title('Grfico de Controle de Custos')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
