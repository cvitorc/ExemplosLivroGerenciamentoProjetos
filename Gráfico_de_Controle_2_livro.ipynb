```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de exemplo (tempo planejado e tempo real gasto em cada fase do projeto)
fases = ['Fundao', 'Estrutura', 'Acabamento']
tempo_planejado = [10, 15, 20]  # Tempo planejado (em semanas)
tempo_real = [9, 16, 19]  # Tempo real gasto (em semanas)

# Calcular diferena entre o tempo planejado e o tempo real
diferenca_tempo = [tempo_real[i] - tempo_planejado[i] for i in range(len(fases))]

# Calcular a mdia e o desvio padro das diferenas
media = np.mean(diferenca_tempo)
desvio_padrao = np.std(diferenca_tempo)

# Definir os limites de controle (3 desvios padro)
limite_superior = media + 3 * desvio_padrao
limite_inferior = media - 3 * desvio_padrao

# Plotar o grfico de controle
plt.figure(figsize=(10, 6))
plt.plot(fases, diferenca_tempo, marker='o', color='blue', linestyle='-')
plt.axhline(media, color='red', linestyle='--', label='Mdia')
plt.axhline(limite_superior, color='green', linestyle='--', label='Limite Superior')
plt.axhline(limite_inferior, color='green', linestyle='--', label='Limite Inferior')
plt.xlabel('Fase do Projeto')
plt.ylabel('Diferena de Tempo (semanas)')
plt.title('Grfico de Controle de Tempo do Projeto')
plt.legend()
plt.grid(True)
plt.show()
```
